```python
import casadi as ca
import numpy as np
```

```python
mu = ca.SX.sym('mu', 2)
sigma = ca.SX.sym('sigma', 1)
sigma_p = ca.SX.sym('sigma_p', 1)
sigma_bar = ca.SX.sym('sigma_bar', 1)

gamma0 = 1
gamma1 = 1
gamma2 = 1
gamma3 = 1
gamma4 = 1
gamma5 = 1
gamma6 = 0.5


df1 = (gamma1/(gamma0*sigma+gamma1)**2-sigma_p/(gamma2+gamma3*sigma_p*sigma_bar))
df2 = (gamma4*sigma_bar/(gamma5*sigma_p*sigma_bar+gamma4)**2 - sigma_p*gamma6)
w0 = 1 - sigma
w1 = 1 - sigma_p

df = ca.vertcat(df1, df2)
g1 = df + mu
g2 = sigma - sigma_bar
g = ca.vertcat(g1, g2, w0, w1)

f = ca.dot(mu, ca.vertcat(w0, w1))
```

```python
sigmas = ca.vertcat(sigma, sigma_bar, sigma_p)
x = ca.vertcat(*[sigmas, mu])
lbg = np.zeros(5)
ubg = ca.vertcat(0, 0, 0, ca.inf, ca.inf)
s_opts = {'ipopt': {'print_level' : 5}}
prob = {'x': x, 'f': f, 'g': g}
lbx = np.zeros(5)
solver = ca.nlpsol('solver', 'ipopt', prob, s_opts)

sol = solver(lbx = lbx, lbg = lbg, ubg = ubg)
print("Solved")
x_out = np.array(sol['x']).flatten()
print(x_out[1:3])
```
